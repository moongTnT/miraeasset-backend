```python
import pandas as pd

from data.fetch_data import fetch_theme_info, fetch_pdf_info, fetch_data_from_db

os_theme_info = pd.DataFrame(fetch_theme_info())

os_theme_info.head()
```

```python
import numpy as np
import math

for i, row in os_theme_info.iterrows():
    etf_tkr, etf_name, theme_kr, ytd, pdf_cnt, top_stk_name, theme_class = row
    
    if math.isnan(ytd):
        ytd="NULL"

    pdf_cnt = pd.DataFrame(fetch_pdf_info(etf_tkr=etf_tkr)).shape[0]
    
    sql = f"""
    INSERT INTO os_theme_info(
        etf_tkr, etf_name,
        theme_kr,
        ytd,
        pdf_cnt,
        top_stk_name,
        theme_class
        )
    VALUES("{etf_tkr}", "{etf_name}", "{theme_kr}", {ytd}, {pdf_cnt}, "{top_stk_name}", "{theme_class}")
    ON DUPLICATE KEY UPDATE 
        etf_tkr="{etf_tkr}",
        etf_name="{etf_name}", 
        theme_kr="{theme_kr}",
        ytd={ytd},
        pdf_cnt={pdf_cnt},
        top_stk_name="{top_stk_name}", 
        theme_class="{theme_class}"
    """
    try:
        fetch_data_from_db(query=sql)
    except:
        print(type(ytd), ytd)
        print(sql)
        break
```
